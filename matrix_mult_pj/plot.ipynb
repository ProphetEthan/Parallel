```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
matrix_size = ['64*64', '512*512', '1024*1024', '2048*2048']
num_process = [1, 2, 4, 8, 16]
speedup_a = [
    [17.4497, 22.1361, 23.3357, 35.9985, 5.97444],
    [771.282, 1337.34, 2127.2, 3471.18, 4168.75],
    [1583.38, 3011.04, 5239.29, 6493.71, 15226.1],
    [6268.74, 9714.98, 23431.3, 43440, 60209.5]
]

speedup_b = [
    [22.3478, 20.3867, 35.217, 23.2653, 4.48408],
    [871.976, 1372.62, 2607.66, 3580.66, 1994.56],
    [1912.78, 3355.74, 4988.79, 10790.5, 17678.3],
    [5878.61, 12376.6, 21935.7, 43440, 66281.7]
]

speedup_ab = [
    [13.2843, 0.128352, 11.0789, 0.0265601, 0.00546426],
    [618.074, 404.086, 35.6908, 38.025, 20.9681],
    [2135.35, 1062.63, 205.744, 129.611, 141.321],
    [5697.76, 4641.14, 2273.03, 1693.86, 1104.17]
]

# 转换数据为 numpy 数组
matrix_size_labels = np.arange(len(matrix_size))
num_process_labels = np.arange(len(num_process))
matrix_size_labels, num_process_labels = np.meshgrid(matrix_size_labels, num_process_labels)

speedup = np.array(speedup_ab)
```

```python
# 创建图形对象
fig = plt.figure(figsize=(14, 6))
ax = fig.add_subplot(111, projection='3d')

# 绘制3D表面图
surf = ax.plot_surface(num_process_labels, matrix_size_labels, speedup.T, cmap='plasma')

# 添加颜色条
fig.colorbar(surf)

# 设置轴标签
ax.set_xlabel('Num Process')
ax.set_ylabel('Matrix Size')
ax.set_zlabel('Speed up Rate')

# 设置轴刻度标签
ax.set_xticks(np.arange(len(num_process)))
ax.set_xticklabels(num_process)
ax.set_yticks(np.arange(len(matrix_size)))
ax.set_yticklabels(matrix_size)

# 显示图形
plt.show()
```

```python
avg_a = np.mean(speedup_a, axis=0)
avg_b = np.mean(speedup_b, axis=0)
avg_ab = np.mean(speedup_ab, axis=0)
```

```python
fig = plt.figure(figsize=(14, 6))
num_processes = [1, 2, 4, 8, 16] 

plt.plot(num_processes, avg_a, label='avg_a')
plt.plot(num_processes, avg_b, label='avg_b')
plt.plot(num_processes, avg_ab, label='avg_ab')

# Adding labels and title
plt.plot(num_processes, avg_a, label='avg_a', marker='o', markersize=8)
plt.plot(num_processes, avg_b, label='avg_b', marker='s', markersize=8)
plt.plot(num_processes, avg_ab, label='avg_ab', marker='^', markersize=8)

# Adding labels and title
plt.xlabel('Number of Processes')
plt.ylabel('Average Speedup')
plt.title('Average Speedup vs. Number of Processes')

# Adding a legend
plt.legend()

# Displaying the plot
plt.show()
```
